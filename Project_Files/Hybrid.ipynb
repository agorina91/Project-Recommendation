# Simple Genre Based Recommender

```python
# provide top recommendations for each genre 

def build_chart(genre, percentile=0.85):
    df = gen_md[gen_md['genre'] == genre]
    
    qualified = df[(df['stars'].notnull())][['title', 'desc', 'url','stars','genre']]
    qualified = qualified.sort_values('stars', ascending=False).head(250)
    
    return qualified
```

# Content Based Recommender 

## Content based using only descriptions

```python
# build a recommender using the podcast descriptions.

tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(md['desc'])
```

```python
tfidf_matrix.shape
```

```python
# used the TF-IDF Vectorizer, calculating the Dot Product will directly give the Cosine Similarity Score.
# use sklearn's linear_kernel instead of cosine_similarities since it is much faster

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
print(cosine_sim.shape)
print(cosine_sim[0])
```

```python
md = md.reset_index(drop=True)
titles = md[['title', 'genre_tags', 'stars']]
indices = pd.Series(md.index, index=md['title'])
indices.head(5)
```

```python
# We now have a pairwise cosine similarity matrix for all the movies in our dataset.
# The next step is to write a function that returns the 30 most similar movies based on the cosine similarity score.

def get_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    podcast_indices = [i[0] for i in sim_scores]
    return titles.iloc[podcast_indices]
```

```python
# test to get the top 10 movies based on cosine similarity on podcast descriptions
get_recommendations('The Shrink Next Door')
```

## Content based incorporating desc and genres

```python
md['desc_keywords'] = md['desc'].apply(lambda x: x.split(' '))

# giving the genre tags 3x the weight
md['genre_tags_x_3'] = md['genre_tags'].apply(lambda x: x*3)
```

```python
md
```

```python
md['desc_genre'] = md['desc_keywords'] + md['genre_tags_x_3']
md['desc_genre'] = md['desc_genre'].apply(lambda x: ' '.join(x))
```

```python
count = CountVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
count_matrix = count.fit_transform(md['desc_genre'])
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
md = md.reset_index(drop=True)
titles =  md[['title', 'genre_tags', 'stars']]
indices = pd.Series(md.index, index=md['title'])
```

```python
get_recommendations('The Shrink Next Door')
```

```python
# results are much better than the description-only recommender, by spitting out items with similar genres
```

### sorted by ratings

```python
def improved_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:26]
    podcast_indices = [i[0] for i in sim_scores]
    
    podcast = md.iloc[podcast_indices][['title', 'desc', 'url','stars','genre_tags']]
    podcast = podcast.sort_values('stars', ascending=False).head(10)
    return podcast
```

```python
improved_recommendations('The Shrink Next Door')
```

```python
# content based engine suffers from some severe limitations.

# - It is only capable of suggesting podcasts which are close to a certain podcast. 
# - the above algoritm does not cate to user's personal preference 
# - highly dependent on what genre tags the podcast is labelled with
```

# Using SVD

```python
from surprise import Dataset, Reader
from surprise import SVD
from surprise import accuracy
from surprise.model_selection import cross_validate, train_test_split
```

```python
# for Surprise, we only need three columns from the dataset
data = df_w_dummies[['user_id', 'podcast_id', 'stars']]
reader = Reader(line_format='user item rating', sep=',')
data = Dataset.load_from_df(data, reader=reader)

# train-test-split
trainset, testset = train_test_split(data, test_size=.2)
```

```python
# instantiate SVD and fit the trainset
svd = SVD()
svd.fit(trainset)
```

```python
predictions = svd.test(testset)
accuracy.rmse(predictions)
accuracy.mae(predictions)
```

```python
svd.predict(1, 172)
```

```python
# taking a look at the first 3 rows of our test set
predictions[:3]
```

```python
# running randomized search on SVD
from surprise.model_selection import RandomizedSearchCV

params = {'n_factors': [1,5,10,25,50,75,100], 'reg_all': [.005,.01,.02,.05,.1,.25,.5], 'lr_all':[.005,.01,.02,.05,.1,.25,.5],
         'n_epochs': [10,20,50]}
model = RandomizedSearchCV(SVD,param_distributions=params,n_jobs=-1,joblib_verbose=10, cv=5)
model.fit(data)
```

```python
# best params and score
print(model.best_score)
print(model.best_params)
```

```python
# now checking best params specifically on test set
better = SVD(n_factors=5,reg_all=.005,lr_all=.25,n_epochs=50)
better.fit(trainset)
better_preds = better.test(testset)
accuracy.rmse(better_preds)
accuracy.mae(better_preds)
```

```python
# seems like the randomizedSearch params did not do better than the baseline model
```

# Hybrid Model incorporating Surprise

```python
podcast_id = md.loc[md.title=='The Shrink Next Door']['podcast_id']
podcast_id[0]
```

```python
def hybrid(user_id, title):
    idx = indices[title]
    podcast_id = md.loc[md.title==title]['podcast_id']
    sim_scores = list(enumerate(cosine_sim[int(idx)]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:26]
    podcast_indices = [i[0] for i in sim_scores]
    podcast = md.iloc[podcast_indices][['title', 'desc', 'url','stars','genre_tags']]
    podcast['est'] = podcast.index.map(lambda x: svd.predict(user_id, md.loc[x]['podcast_id']).est)
    podcast = podcast.sort_values('est', ascending=False)
    return podcast.head(10)
```
