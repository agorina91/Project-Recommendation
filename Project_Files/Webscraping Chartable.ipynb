```python
from bs4 import BeautifulSoup as BS
import requests
import pandas as pd
```

```python
# using bs4 to get links to each podcast
def get_links():
    links = []
    for i in range(1,11):
        url = 'https://chartable.com/charts/itunes/us-all-podcasts-podcasts?page='+str(i)
        r = requests.get(url)
        soup = BS(r.content, 'html.parser')
        link_per_page = []
        for i in soup.find_all(class_='link blue',href=True):
            link_per_page.append(i['href'])
            links_50perpage = link_per_page[0:50]
        links.append(links_50perpage)
    return links

all_links = get_links()
```

```python
# flatten nested list
import operator
import functools 
link_list = functools.reduce(operator.add, all_links)
len(link_list), link_list
```

```python
# remove all duplicate links
# link_list1 = list(set(link_list))
link_list1 = sorted(set(link_list), key=link_list.index)
len(link_list1), link_list1 
```

```python
# clear out miscellaneous links
cleaned_links = []
for i in link_list1:
    if 'podcasts' in i:
        cleaned_links.append(i)
```

```python
print(len(cleaned_links))
cleaned_links
```

```python
# Using bs4 to get title for all podcasts
title_list = []
for i in cleaned_links:
    url = i
    r = requests.get(i)
    soup = BS(r.content, 'html.parser')
    title = soup.find_all(class_='f1-ns f2 fw1 mb2 dark-blue')
    for i in range(len(title)):
        title_list.append({'url':url, 'title':title[i].get_text()})
title_list
```

```python
len(title_list)
```

```python
df_title = pd.DataFrame(title_list)
df_title.head()
```

```python
df_title.to_csv("podcast_titles.csv", index=False)
```

```python
# Using bs4 to get descriptions for all podcasts
descriptions = []
for i in cleaned_links:
    url = i
    time.sleep(1)
    r = requests.get(i)
    soup = BS(r.content, 'html.parser')
    summary = soup.find_all(class_='almost-silver')
    for i in range(len(summary)):
        descriptions.append({'url':url, 'desc': summary[i].get_text()})
descriptions
```

```python
len(descriptions)
```

```python
df_desc = pd.DataFrame(descriptions)
df_desc.head()
```

```python
df_desc.to_csv("podcast_descriptions.csv", index=False)
```

```python
# Using bs4 to get genres for all podcasts
genres = []
for i in cleaned_links:
    url = i
    time.sleep(1)
    r = requests.get(url)
    soup = BS(r.content, 'html.parser') 
    genre = []
    for i in soup.find_all(class_='link blue',href=True):
        if 'genre' in i['href']:
            genre.append(i.get_text())
    genres.append({'url':url, 'genre': genre})
genres
```

```python
len(genres)
```

```python
df_genre = pd.DataFrame(genres)
df_genre.head()
```

```python
df_genre.to_csv("podcast_genres.csv", index=False)
```

```python
# Using bs4 to get most recent 50 reviews for all podcasts

all_reviews = []
for i in cleaned_links:
    url = i+'/reviews/page/1'
    try:
        r = requests.get(url)
        print('requesting:', url)
    except:
        time.sleep(1)
        r = requests.get(url)
        print('requesting:', url)
    soup = BS(r.content, 'html.parser')
    reviews = soup.find_all(class_='mb4')
        
    stars = []
    for j in reviews:
        star = j.findAll(class_='dib full icon-star')
        star_rating = []
        for k in range(len(star)):
            star_rating.append(star[k].get_text())
        stars.append(star_rating)
    star_count = []
    for i in stars:
        star_count.append(i.count('☆'))

    users = []
    for j in reviews:
        user = j.find(class_='f6 mt1 silver').text
        users.append(user.split(' via')[0].replace('\n',''))
        
    reviews_summ = []
    for i in reviews:
        review = i.findAll(class_='f5 lh-copy mt2')
        all_review = []
        for k in range(len(review)):
            all_review.append(review[k].get_text())
        reviews_summ.extend(all_review)

    
    all_reviews.append({'url': url, 'stars': star_count, 'user': users, 'reviews': reviews_summ})

all_reviews
```
